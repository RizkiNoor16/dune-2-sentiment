## Clean Dataset

```python
import pandas as pd
data = pd.read_csv("dune-16-03-2024:20.45.csv")
data
```

```python
data['Star'].value_counts()
```

```python
data['Lang'].value_counts()
```

```python
# Decide to use only english review
data = data.loc[data['Lang'] == 'en']
data
```

```python
data['label'] = data['Star'] -1
```

```python
data = data.drop(columns=['Unnamed: 0', 'Author', 'Date', 'Lang','Star'])
data
```

```python
from sklearn.model_selection import train_test_split


# Pembagian data menjadi train (80%) dan temp (20%)
train, val = train_test_split(data, test_size=0.2, random_state=42)


# Periksa jumlah baris di setiap bagian
print("Jumlah data pelatihan:", len(train))
print("Jumlah data validasi:", len(val))
```

```python
train.to_csv("dataset/train.csv")
val.to_csv("dataset/val.csv")
```
